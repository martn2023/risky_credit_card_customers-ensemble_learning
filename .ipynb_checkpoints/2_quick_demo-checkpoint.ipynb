# II. Bringing in (Competing Models) for Classification

### 1: Naming and Importing the Tools
- Logistic Regression (chapter 4)
- SVC (chapter 5, but skipped)
- Random Forest (continuation of decision trees in chapter 6)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

logistic_regression_classifier = LogisticRegression()
svm_classifier = SVC()
random_forest_classifier = RandomForestClassifier()
```

### 2: Naming the Judge/Foreman
- VotingClassifier

```python
from sklearn.ensemble import VotingClassifier

voting_clf = VotingClassifier (estimators=[('lr', logistic_regression_classifier), ('rf', random_forest_classifier), ('svc', svm_classifier)], voting='hard')

voting_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

for clf in (logistic_regression_classifier, random_forest_classifier, svm_classifier, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

### 3: Accuracy Evaluation

When I was the VP of Finance at a Y Combinator startup, we had a debate internally while trying to build composite consumer profiles out of multiple data sources. Assume that data source A was 80% accurate and that data source B was 60% accurate.
- One founder believed that combining data from multiple sources would give us the overall most accurate picture (ensemble logic here)
- The other founder believed that weaker data sources would dilute the stronger data sources, so the end result would fall from 80% to something between 60% and 80%.

Both sound reasonable in a vacuum. When we ran the experiment here though with different models, the voting classifier achieved the highest accuracy at **98%** (individual models ranged from 86.7% to 97.7%).
